```python
import pandas as pd
import benchmax.evaluation as ev

def get_data(solvers):
    param = {}
    for solver in solvers:
        param["stats_smtrat_15m_6Gi_"+solver+".xml"] = {"smtrat-static": solver, "smtrat-"+solver: solver}
    return ev.xmls_to_pandas(param,['mcsat-explanation-nlsat_explanation_called','mcsat-explanation-onecellcad_explanation_called','mcsat-explanation-fm_explanation_called'])

df = get_data(["MCSATNL","MCSATOCNNASC","MCSATOCNNDSC","MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13","MCSATOCLWH22","MCSATOCLWH33","MCSATFMICPVSNL","MCSATFMICPVSOCNNASC","MCSATFMICPVSOCLWH11","MCSATFMICPVSOCLWH12","MCSATFMICPVSOCLWH13"])


for solver in ["MCSATNL"]:
    df.loc[:,(solver,'cells_constructed')] = df[(solver,'mcsat-explanation-nlsat_explanation_called')].fillna(0).astype(int)
for solver in ["MCSATOCNNASC","MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13","MCSATOCLWH22","MCSATOCLWH33"]:
    df.loc[:,(solver,'cells_constructed')] = df[(solver,'mcsat-explanation-onecellcad_explanation_called')].fillna(0).astype(int) 
for solver in ["MCSATFMICPVSNL","MCSATFMICPVSOCNNASC","MCSATFMICPVSOCLWH11","MCSATFMICPVSOCLWH12","MCSATFMICPVSOCLWH13"]:
    df.loc[:,(solver,'cells_constructed')] = df[(solver,'mcsat-explanation-fm_explanation_called')].fillna(0).astype(int)

for solver in ["MCSATNL", "MCSATFMICPVSNL"]:
    df.loc[:,(solver,'oc_called')] = 0
for solver in ["MCSATOCNNASC","MCSATOCNNDSC","MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13","MCSATOCLWH22","MCSATOCLWH33"] + ["MCSATFMICPVSOCNNASC","MCSATFMICPVSOCLWH11","MCSATFMICPVSOCLWH12","MCSATFMICPVSOCLWH13"]:
    df.loc[:,(solver,'oc_called')] = df[(solver,'mcsat-explanation-onecellcad_explanation_called')].fillna(0).astype(int)

for solver in ["MCSATNL", "MCSATFMICPVSNL"]+["MCSATOCNNASC","MCSATOCNNDSC","MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13","MCSATOCLWH22","MCSATOCLWH33"] + ["MCSATFMICPVSOCNNASC","MCSATFMICPVSOCLWH11","MCSATFMICPVSOCLWH12","MCSATFMICPVSOCLWH13"]:
    df.loc[:,(solver,'nl_called')] = df[(solver,'mcsat-explanation-nlsat_explanation_called')].fillna(0).astype(int)

df = df.join(ev.virtual_best(df, ["MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13"], "VB", ['oc_called','cells_constructed','nl_called']))
df = df.join(ev.virtual_best(df, ["MCSATFMICPVSOCLWH11","MCSATFMICPVSOCLWH12","MCSATFMICPVSOCLWH13"], "VB+", ['oc_called','cells_constructed','nl_called']))
df
```

```python
def get_or_zero(data,key):
    if key in data:
        return data[key]
    else:
        return 0

solvers = ["MCSATOCLWH11","MCSATFMICPVSOCLWH11","VB","VB+"]
labels = ["LW-EQ-BC","LW-EQ-BC+","VB","VB+"]
data = []
idx = []

idx.append("solved")
data.append([str(df[(solver,'answer')].value_counts()['sat'] + df[(solver,'answer')].value_counts()['unsat']) for solver in solvers])

idx.append("mean runtime")
data.append([str(round(ev.filter_solved(df,solver)[(solver,'runtime')].mean(),2)) for solver in solvers])

idx.append("explanation called")
data.append([str(ev.filter_solved(df,solver)[(solver,'cells_constructed')].sum()) for solver in solvers])

idx.append("oc called")
data.append([str(ev.filter_solved(df,solver)[(solver,'oc_called')].sum()) + " ("+ str(round(ev.filter_solved(df,solver)[(solver,'oc_called')].sum()/ev.filter_solved(df,solver)[(solver,'cells_constructed')].sum()*100, 2)) +"%)" for solver in solvers])

idx.append("nl called")
data.append([str(ev.filter_solved(df,solver)[(solver,'nl_called')].sum()) + " ("+ str(round(ev.filter_solved(df,solver)[(solver,'nl_called')].sum()/ev.filter_solved(df,solver)[(solver,'cells_constructed')].sum()*100, 2)) +"%)" for solver in solvers])

def filter_solved_oc(df, solver):
    return ev.filter_solved(df,solver)[df[(solver,'oc_called')]>0]

idx.append("solved with sc")
data.append([str(len(filter_solved_oc(df,solver))) for solver in solvers])

idx.append("mean runtime with sc")
data.append([str(round(filter_solved_oc(df,solver)[(solver,'runtime')].mean(),2)) for solver in solvers])

idx.append("explanation called with sc")
data.append([str(filter_solved_oc(df,solver)[(solver,'cells_constructed')].sum()) for solver in solvers])

idx.append("oc called with sc")
data.append([str(filter_solved_oc(df,solver)[(solver,'oc_called')].sum()) + " ("+ str(round(filter_solved_oc(df,solver)[(solver,'oc_called')].sum()/filter_solved_oc(df,solver)[(solver,'cells_constructed')].sum()*100, 2)) +"%)" for solver in solvers])

idx.append("nl called with sc")
data.append([str(filter_solved_oc(df,solver)[(solver,'nl_called')].sum()) + " ("+ str(round(filter_solved_oc(df,solver)[(solver,'nl_called')].sum()/filter_solved_oc(df,solver)[(solver,'cells_constructed')].sum()*100, 2)) +"%)" for solver in solvers])


summary = pd.DataFrame(data, index=idx,columns=labels)

summary
```
