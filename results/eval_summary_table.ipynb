```python
import pandas as pd
import benchmax.evaluation as ev

def get_data(solvers):
    param = {}
    for solver in solvers:
        param["stats_smtrat_15m_6Gi_"+solver+".xml"] = {"smtrat-static": solver, "smtrat-"+solver: solver}
    return ev.xmls_to_pandas(param,['mcsat-explanation-nlsat_explanation_called','mcsat-explanation-onecellcad_explanation_called','mcsat-explanation-fm_explanation_called'])

df = get_data(["MCSATNL","MCSATOCNNASC","MCSATOCNNDSC","MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13","MCSATOCLWH22","MCSATOCLWH33","MCSATFMICPVSNL","MCSATFMICPVSOCNNASC","MCSATFMICPVSOCLWH11","MCSATFMICPVSOCLWH12","MCSATFMICPVSOCLWH13"])

for solver in ["MCSATNL", "MCSATFMICPVSNL"]:
    df.loc[:,(solver,'oc_called')] = 0

for solver in ["MCSATOCNNASC","MCSATOCNNDSC","MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13","MCSATOCLWH22","MCSATOCLWH33"]:
    df.loc[:,(solver,'oc_called')] = df[(solver,'mcsat-explanation-onecellcad_explanation_called')].fillna(0).astype(int)
    
for solver in ["MCSATFMICPVSOCNNASC","MCSATFMICPVSOCLWH11","MCSATFMICPVSOCLWH12","MCSATFMICPVSOCLWH13"]:
    df.loc[:,(solver,'oc_called')] = df[(solver,'mcsat-explanation-onecellcad_explanation_called')].fillna(0).astype(int)

df = df.join(ev.virtual_best(df, ["MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13"], "VB-LW", ['oc_called']))
df = df.join(ev.virtual_best(df, ["MCSATFMICPVSOCLWH11","MCSATFMICPVSOCLWH12","MCSATFMICPVSOCLWH13"], "VB-LW+", ['oc_called']))
df = df.join(ev.virtual_best(df, ["MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13","MCSATOCNNASC"], "VB", ['oc_called']))
df = df.join(ev.virtual_best(df, ["MCSATFMICPVSOCLWH11","MCSATFMICPVSOCLWH12","MCSATFMICPVSOCLWH13","MCSATFMICPVSOCNNASC"], "VB+", ['oc_called']))

df.loc[:,'set'] = df.index.str.split(pat='/').map(lambda x: x[0])
df
```

```python
def get_or_zero(data,key):
    if key in data:
        return data[key]
    else:
        return 0

solvers = ["MCSATNL", "MCSATOCNNASC", "MCSATOCNNDSC","MCSATOCLWH11","MCSATOCLWH12","MCSATOCLWH13", "VB-LW","VB","MCSATOCLWH22","MCSATOCLWH33", "MCSATFMICPVSNL", "MCSATFMICPVSOCNNASC","MCSATFMICPVSOCLWH11","MCSATFMICPVSOCLWH12","MCSATFMICPVSOCLWH13", "VB-LW+", "VB+"]
labels = ["NL","OC-ASC", "OC-DSC","LW-EQ-BC","LW-EQ-CH","LW-EQ-LDB","VB-LW","VB","LW-CH-CH","LW-LDB-LDB","NL+","OC-ASC+","LW-EQ-BC+","LW-EQ-CH+","LW-EQ-LDB+", "VB-LW+","VB+"]
data = []

for solver in solvers:
    data.append(tuple([
        df[(solver,'answer')].value_counts()['sat'] + df[(solver,'answer')].value_counts()['unsat'],
        df[(solver,'answer')].value_counts()['sat'],
        df[(solver,'answer')].value_counts()['unsat'],
        get_or_zero(df[df[(solver,'oc_called')]>0][(solver,'answer')].value_counts(),'sat') + get_or_zero(df[df[(solver,'oc_called')]>0][(solver,'answer')].value_counts(),'unsat'),
        get_or_zero(df[df[(solver,'oc_called')]==0][(solver,'answer')].value_counts(),'sat') + get_or_zero(df[df[(solver,'oc_called')]==0][(solver,'answer')].value_counts(),'unsat')
    ]))

data.append([11552,5069,5379,0,0])

summary = pd.DataFrame(data, index=labels+['total'],columns=["solved", "sat solved", "unsat solved", "solved with single cell", "solved without single cell"])

summary
```
